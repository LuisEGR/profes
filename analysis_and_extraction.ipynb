```python
def getRealNameFromAlias(clase):
    if clase == None:
        return ""
    clase = clase.upper()
    if clase in alias:
         return alias[clase];
    return clase

#test
real = getRealNameFromAlias("POO")
real
```

```python
def similar(a, b):
    if a != None and b != None:
        return SequenceMatcher(None, a.upper(), b.upper()).ratio()
    return 0

#test
similar("hola", "Hola")
```

```python
def getMateriaNormalized(c):
    if c != None and c != "---": 
        c = c.strip()
        c_m = c.upper()
        c_al = getRealNameFromAlias(c)
        if c_m != c_al:
            best_choice = c_al
            best_percent = 1        
        else:
            best_choice = c
            best_percent = 0
            for mat in materias:                        
                similitud = similar(mat, c)
                if(similitud > 0.5 and similitud > best_percent):
                    best_percent = similitud
                    best_choice = mat

        if best_percent != 0: 
            return best_choice
        else:
            return False
```

```python
def getMaterias():
    for c in data_existent['clase']:        
        if c != None and c != "---": 
            c = c.strip()
            c_m = c.upper()
            c_al = getRealNameFromAlias(c)
            if c_m != c_al:
                best_choice = c_al
                best_percent = 1
            else:
                best_choice = c
                best_percent = 0
                for mat in materias:                        
                    similitud = similar(mat, c)
                    if(similitud > 0.5 and similitud > best_percent):
                        best_percent = similitud
                        best_choice = mat
                        
            if best_percent != 0:                 
                print(best_percent)   
                print(c + "->" + best_choice)
            print( "," + best_choice)
            
getMaterias()
```

```python
import json
all_items = list()
for row in data_existent.iterrows():
    r = row[1]
    mat = getMateriaNormalized(r['clase'])        
    if mat and r['comentario']:
        prof_d = {
            "profesor": r['profesor'],
            "materia": mat,
            "comentario": r['comentario'].replace("'", '"'),
            "f": r['f'],
            "a": r['a'],
            "c": r['c'],
            "recomendado": r['recomendado']
        }
        all_items.append(prof_d)
#print(all_items)
with open("eval_normal.json", 'w') as outfile:
    json.dump(all_items, outfile)
```
